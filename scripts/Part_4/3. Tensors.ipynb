### One of the most common error in deep learning: shape errors
There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match
* `(3, 2) @ (3, 2)` This won't work
* `(3, 2) @ (2, 3)` Will work
* `(2, 3) @ (3, 2)` Will work
2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# Shapes for matrix multiplÃ¬cation
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul (it's an alias for matmul)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues we can manipulate the shape of one of our tensors using a transpose.

A transpose switches the axes or the dimensions of a given tensor.


```python
tensor_B, tensor_B.T
```

## Reshaping, stacking, squeezing and unsqueezing tensor
* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - Combine multiple tensor on top of each other or side by side
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - Adds a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x cause a view of a tensor shares the same memory as the original tensor-
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim = 0)
x_stacked, x_stacked.shape
```

```python
# torch.squeeze
x_reshaped, x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
# Torch.permute - rearranges the dimensions of the target tensor in a specified tensor
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axes (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shift axes 0 to 1, 1 to 2 and 2 to 0
x_permuted.shape
```

```python
x_original
```

```python
x_original[0, 0, 0] = 999
```

```python
x_permuted[0, 0, 0]
```

## PyTorch tensors & Numpy
Numpy is a popular scientific Python numerical computing library.
And because of this, PyTorch has functionallity to interact with it.
* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor to NumPy -> torch.Tensor.numpy()

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Tensor to NumPy
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

### 3. Putting tensors and models on the GPU

The reason we want tensors/models on the GPU is because using a GPU results in faster computations
